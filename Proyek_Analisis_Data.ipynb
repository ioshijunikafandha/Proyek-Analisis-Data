```
# Menggunakan metode read_csv dari pandas untuk membaca file CSV dari URL yang diberikan
# dan menyimpannya dalam dataframe hour_df
hour_df = pd.read_csv("https://raw.githubusercontent.com/ioshijunikafandha/Proyek-Analisis-Data/master/data/hour.csv")

# Menampilkan lima baris pertama dari DataFrame hour_df
hour_df.head()
```

### Cleaning Data

Membersihkan data pada dataframe hour

```
# Mengganti tipe data yang salah pada dataframe hour
datetime_columns = ["dteday"]

for column in datetime_columns:
  hour_df[column] = pd.to_datetime(hour_df[column])

# Menampilkan informasi tentang struktur DataFrame setelah perubahan tipe data
hour_df.info()
```

Membersihkan data pada dataframe day

```
# Mengganti tipe data yang salah pada dataframe day
datetime_columns = ["dteday"]

for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])

# Menampilkan informasi tentang struktur DataFrame setelah perubahan tipe data
day_df.info()
```

## Exploring Data Analysis (EDA)

```
# Menggunakan metode describe() dengan argumen include='all',
# untuk memberikan ringkasan statistik deskriptif untuk semua kolom
day_df.describe(include='all')
```

```
# Menggunakan metode describe() dengan argumen include='all',
# untuk memberikan ringkasan statistik deskriptif untuk semua kolom
hour_df.describe(include="all")
```

```
# Membuat kamus season_label untuk memetakan kode musim (dalam kolom 'season') ke label musim yang sesuai.
# Menggunakan metode .map()
season_label = {
    1: 'Summer',
    2: 'Autumn',
    3: 'Spring',
    4: 'Winter'
}
day_df['season_label'] = day_df['season'].map(season_label)
```

```
# Melakukan pengelompokkan berdasarkan label musim yang telah dibuat sebelumnya ('season_label')
# Kemudian menghitung rata-rata nilai 'cnt' (jumlah sepeda yang disewa) untuk setiap label musim,
# dan mengurutkan hasil berdasarkan nilai 'cnt' secara menaik.
day_df.groupby('season_label')['cnt'].mean().reset_index().sort_values("cnt")
```

```
# Membuat kamus weekday_labels untuk memetakan kode hari dalam seminggu ke label hari yang sesuai.
# Menggunakan metode .map()
weekday_label = {
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}
day_df['weekday_label'] = day_df['weekday'].map(weekday_label)
```

```
# Melakukan pengelompokkan berdasarkan label hari yang telah dibuat sebelumnya ('weekday_label')
# Kemudian menghitung rata-rata nilai 'cnt' (jumlah sepeda yang disewa) untuk setiap label hari,
# dan mengurutkan hasil berdasarkan nilai 'cnt' secara menaik.
day_df.groupby('weekday_label')['cnt'].mean().reset_index().sort_values("cnt")
```

## Visualisasi Data

```
# Melakukan pengelompokan berdasarkan label musim dan menghitung rata-rata jumlah pengguna sepeda
avg_season = day_df.groupby('season_label')['cnt'].mean().reset_index()

# Mengatur ukuran gambar plot
plt.figure(figsize=(10, 7))

# Membuat horizontal bar chart
plt.barh(avg_season['season_label'], avg_season['cnt'], color=('#432818', '#582f0e', '#eddea4', '#685634'))

# Menambahkan judul dan label sumbu
plt.title('Rata-rata Jumlah Pengguna Sepeda berdasarkan Musim')
plt.xlabel('Rata-rata Jumlah Pengguna Sepeda')
plt.ylabel('Musim')

# Memutar label sumbu y sebesar 45 derajat
plt.yticks(rotation=45)

# Menambahkan grid pada sumbu x
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Menampilkan plot
plt.show()
```

```
# Menghitung rata-rata penyewaan sepeda berdasarkan hari dalam seminggu
avg_weekday = day_df.groupby('weekday')['cnt'].mean()

# Label untuk pie donat chart pada hari dalam seminggu
labels_weekday = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']

# Pie donat chart untuk rata-rata pada setiap hari
plt.pie(avg_weekday, labels=labels_weekday, autopct='%1.1f%%', colors=('#7f4f24', '#eddea4', '#432818', '#f6f4d2'), wedgeprops=dict(width=0.4))
plt.title('Rata-rata Penyewaan Sepeda pada Hari dalam Seminggu')

# Menampilkan plot
plt.show()
```
